```python
#Compute IQR for all numeric columns at once
Q1 = Bhouse.quantile(0.25)
Q3 = Bhouse.quantile(0.75)

IQR = Q3 - Q1

print(IQR)
```

```python
#Get lower and upper bounds for all columns

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
#Detect outliers across all columns
outliers = ((Bhouse < lower_bound) | (Bhouse > upper_bound))

outliers.sum().sort_values(ascending=False)
```

```python
#4️⃣ Cap outliers for all columns at once (IQR capping)
Bhouse_capped = Bhouse.clip(lower=lower_bound, upper=upper_bound, axis=1)
```
